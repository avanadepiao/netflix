#### 1. 概要
Matplotlibライブラリはグラフを可視化するためのモジュールが含まれている

```python
# module import
import numpy as np
import matplotlib.pyplot as plt
# notebook内でグラフを表示する ←挙動がおかしい...
%matplotlib inline
```

#### 2. 基本操作
 `plt.plot(listA)`
 `plt.show()`

```python
# 1 gragh
d = [0, 2, 5, 9, 14, 20]

# x: key y:value
plt.plot(d)
plt.show()
```

```python
# ndarrayを作成
x = [0, 1, 2, 3, 4]
y = [0, 3, 6, 9, 12]
aryx = np.array(x)
aryy = np.array(y)

# plot関数で描画
plt.plot(aryx, aryy)
plt.show()
```

```python
# 複数のグラフ
data = [0, 1, 4, 9, 16]
x = [0, 1, 2, 3, 4]
y = [0, 1, 2, 3, 4]

# plot
plt.plot(x, y)
plt.plot(data)
plt.show()
```

```python
# style
data = [0, 1, 4, 9, 16]
x = [0, 1, 2, 3, 4]
y = [0, 1, 2, 3, 4]

# linestyle, color, marker
plt.plot(x, y, linestyle='--', color='blue', marker='o')
plt.plot(data, linestyle=':', color='green', marker='*')
plt.show()
```

```python
# legende()で凡例を表示
data = [0, 1, 4, 9, 16]
x = [0, 1, 2, 3, 4]
y = [0, 1, 2, 3, 4]

#plot
plt.plot(x, y, linestyle='--', color='blue', marker='o', label='linear')
plt.plot(data, linestyle=':', color='green', marker='*', label='quad')
plt.legend()
plt.show()
```

```python
# 曲線
x = np.arange(-np.pi, np.pi, 0.1)

# cos, sin
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x))

plt.title('cos and sin Curves')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# 直線 プロット数を少なくすると、曲線 → 直線
x = np.arange(-np.pi, np.pi, 0.5)
# cos, sin
plt.plot(x, np.cos(x), marker='o')
plt.plot(x, np.sin(x), marker='o')

plt.title('cos and sin Curves')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_exp(title, xlabel, ylabel, label,grid):
    
    x = np.arange(-2, 2, 0.1)
    y = np.exp(x)
    plt.plot(x, y, label=label)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(grid)
    plt.legend()
    plt.show()

plot_exp('exponential function', 'x', 'y', 'Exponential', True)
```

#### 3. 散布図scatter
`plt.scatter(lista, listb)`

```python
x = np.random.rand(20)
y = np.random.rand(20)

# marker size, 透明度
plt.scatter(x, y, s=100, alpha=0.5)
plt.show()
```

```python
plt.plot(x, y, '*', color='blue')
plt.show()
```

```python
x = np.arange(1, 11, 1)
y = np.random.rand(10)
plt.bar(x, y)
plt.show()
```

```python
# 正規分布に基づく1000個の数値の要素からなる配列
d = np.random.randn(1000)
plt.hist(d, bins=20)
plt.show()
```

```python
# 10行10列のランダム要素からなる行列
ary1 = np.random.rand(100)
ary2 = ary1.reshape(10, 10)

# ヒートマップを描画
im = plt.imshow(ary2)
plt.colorbar(im)
plt.show()
```
